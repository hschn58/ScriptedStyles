```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Set the range for the axes
rang = 10

# Create a figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Set the limits for the 3D plot
ax.set_xlim3d(-rang, rang)
ax.set_ylim3d(-rang, rang)
ax.set_zlim3d(-rang, rang)

# Define the size and scale
granularity = 10**2
scale = 1 / granularity


# Define the Gaussian function
def gaussian(x, y, b, a, xval, yval):
    return a*np.exp(-((x-xval*4)**2 + (y-yval*4)**2)**(1/2)**1.01)

# Generate a range of x values
x = np.arange(start=-rang, stop=rang, step=scale)

# Create a meshgrid for x and y values
x, y = np.meshgrid(x, x)

# Calculate the Gaussian function output for b=0
out1 = gaussian(x, y, 0.5*rang, 4, xval=1, yval=0)
out2 = gaussian(x, y, 0.5*rang, 4, xval=-1, yval=0)

out3 = gaussian(x, y, 0.5*rang, 4, xval=0, yval=1)
out4 = gaussian(x, y, 0.5*rang, 4, xval=0, yval=-1)

# Plot the surface
#add the outs
ax.plot_surface(x, y, out1+out2+out3+out4, color='w')

# Show the plot
plt.show()
```

```python
import numpy as np

# Define the function to calculate the center of mass
def calculate_com(bitmap):
    # Get the dimensions of the bitmap
    rows, cols = bitmap.shape
    
    # Create arrays of indices for the rows and columns
    y_indices, x_indices = np.indices(bitmap.shape)
    
    # Calculate the total mass (number of 1s)
    total_mass = np.sum(bitmap)
    
    if total_mass == 0:
        return None  # Return None if there are no 1s in the bitmap
    
    # Calculate the center of mass coordinates
    com_x = np.sum(x_indices * bitmap) / total_mass
    com_y = np.sum(y_indices * bitmap) / total_mass
    
    return com_x, com_y

# Example usage
dimension = 10
bitmap = np.zeros((dimension, dimension))

# Create a sample boolean map
bitmap[3, 3] = 1
bitmap[5, 7] = 1
bitmap[8, 2] = 1

# Calculate the center of mass
com = calculate_com(bitmap)
print(f"The center of mass is at: {com}")

# Visualize the bitmap and the center of mass
import matplotlib.pyplot as plt

plt.imshow(bitmap, cmap='gray')
if com:
    plt.scatter(com[0], com[1], color='red')  # Mark the center of mass
plt.title('Boolean Bitmap with Center of Mass')
plt.show()
```

```python
import yfinance as yf
```

```python
key = ['HD', 'HON', 'LMT', 'WSM', 'CLX', 'COST', 'GIS', 'MKC', 'PEP', 'BLK', 'ICE', 'JPM', 'PYPL', 'USB', 'ABT', 
       'AMGN', 'BMY', 'MRK', 'TMO', 'UNH', 'JNJ', 'WM', 'AAPL', 'AMZN', 'META', 'GOOG', 'MSFT', 'ADBE', 'ANET', 
       'CSCO', 'EBAY', 'ORCL', 'TXN', 'CNI', 'UNP', 'UPS', 'NEE', 'DUK', 'AMT', 'DLR', 'O']

# Fetch data from yfinance
data_list = []
for ticker in key:
    stock = yf.Ticker(ticker).history(period='15d')
    data_list.append(stock)
```

```python
import pandas as pd

DF = pd.concat(data_list, axis=1, keys=key)
DF.columns.names = ['Stock Tickers','Stock Info']
Closing_Price = DF.xs(key='Close', axis=1, level='Stock Info')

# Calculate percentage changes
Percent_Change1 = (Closing_Price / Closing_Price.shift(5) - 1) * 100
Percent_Change2 = (Closing_Price / Closing_Price.shift(4) - 1) * 100
Percent_Change3 = (Closing_Price / Closing_Price.shift(3) - 1) * 100

Five_Days = Percent_Change1[Percent_Change1 < -3]
Four_Days = Percent_Change2[Percent_Change2 < -3]
Three_Days = Percent_Change3[Percent_Change3 < -3]
```

```python
def fix_date(df):
    df.reset_index(inplace=True)
    df['Date'] = df['Date'].astype(str).str.split(' ').str[0]
```

```python
# Round to 3 decimal places
for ticker in key:
    Five_Days[ticker] = Five_Days[ticker].round(3)
    Four_Days[ticker] = Four_Days[ticker].round(3)
    Three_Days[ticker] = Three_Days[ticker].round(3)

# Fix date format
fix_date(Five_Days)
fix_date(Four_Days)
fix_date(Three_Days)
```

```python
import os
output_dir = os.getcwd()
output_dir
```

```python

five_days_path = os.path.join(output_dir, 'Five_Days.csv')
five_days_last_path = os.path.join(output_dir, 'Five_Days_Last.csv')

four_days_path = os.path.join(output_dir, 'Four_Days.csv')
four_days_last_path = os.path.join(output_dir, 'Four_Days_last.csv')

three_days_path = os.path.join(output_dir, 'Three_Days.csv')
three_days_last_path = os.path.join(output_dir, 'Three_Days_Last.csv')
```

```python
def format_dataframe(df):
    df_copy = df.copy()
    df_copy.fillna(0, inplace=True)
    df_copy = df_copy.applymap(lambda x: int(x) if x == 0 else x)
    return df_copy

Five_Days_formatted = format_dataframe(Five_Days)
Four_Days_formatted = format_dataframe(Four_Days)
Three_Days_formatted = format_dataframe(Three_Days)

Five_Days_formatted.to_csv(five_days_path, index=False)
Five_Days.iloc[-1:].dropna(axis=1).to_csv(five_days_last_path, index=False)

Four_Days_formatted.to_csv(four_days_path, index=False)
Four_Days.iloc[-1:].dropna(axis=1).to_csv(four_days_last_path, index=False)

Three_Days_formatted.to_csv(three_days_path, index=False)
Three_Days.iloc[-1:].dropna(axis=1).to_csv(three_days_last_path, index=False)
```
